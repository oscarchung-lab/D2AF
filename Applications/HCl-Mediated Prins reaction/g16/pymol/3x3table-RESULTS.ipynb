```python
from PIL import Image,ImageFont,ImageDraw
import matplotlib.pyplot as plt

import os
```

```python
COL =3
ROW =3

unit_h = 1200
unit_w = 1200
method_interval = 600
pict_interval = 175

out_put_name = 'concat.png'
save_quality = 1200
method = 3
```

```python

image_list = ['M2_bond_delta','M1_frag','M1_frag_show','M'+str(method)+'_total','M'+str(method)+'_bond','M'+str(method)+'_angle']
numlist = [3,4,5,7,8,9]
text_list = ['','Strain map','Frag show','Strain map','M2 bond','M2 angle']
#RGBA就是把背景改成透明的
canvas = Image.new('RGBA',(unit_w*COL+ pict_interval*(COL-1)+method_interval,(unit_h+pict_interval)*ROW))
pen = ImageDraw.Draw(canvas)
method_font = ImageFont.truetype('arial.ttf', 150)
subtitle_font = ImageFont.truetype('arial.ttf', 100)

#Write head

for lab, ord in enumerate(numlist):
    now_row = (ord-1)//3
    now_column = (ord-1)%3
    print(now_row,now_column)
    if text_list[lab] != '':
        pen.text((method_interval+(now_column+0.3)*unit_w+pict_interval*(now_column),(pict_interval+unit_h)*now_row+50),text_list[lab],fill='black', font=subtitle_font)
    tmp = Image.open(image_list[lab]+'.png')
    canvas.paste(tmp,(method_interval+now_column*(unit_w+pict_interval),(unit_h+pict_interval)*now_row+pict_interval))
pen.text((0,pict_interval*2+unit_h*1.5),'M1',fill='black', font=method_font)
pen.text((0,pict_interval*3+unit_h*2.5),'M2',fill='black', font=method_font)
pen.text((method_interval*1+pict_interval*2+unit_w*(2+0.25),pict_interval*1),'Bond(TS1-R1)',fill='black',font=subtitle_font)
for n in range(1,4):
    pen.line((0,pict_interval*n+unit_h*n-50,unit_w*COL+ pict_interval*(COL-1)+method_interval,pict_interval*n+unit_h*n-20),fill='black',width=13)
'''
#Bond 
tmp = Image.open('M2_bond_delta.png')
canvas.paste(tmp,(unit_w*2 + method_interval+ pict_interval*2,unit_h*0+ pict_interval))
#M1
pen.text((0,pict_interval*2+unit_h*1.5),'M1',fill='black', font=method_font)
pen.text((method_interval+0.3*unit_w,pict_interval+unit_h+20),'Strain map',fill='black', font=subtitle_font)
tmp = Image.open('M1_frag.png')
canvas.paste(tmp,(method_interval,unit_h*1+ pict_interval*2))

pen.text((method_interval+unit_w*(1+0.3)+pict_interval*1,pict_interval+unit_h+20),'Frag show',fill='black', font=subtitle_font)
tmp = Image.open('M1_frag_show.png')
canvas.paste(tmp,(method_interval+unit_w*1+pict_interval*1,unit_h*1+ pict_interval*2))
#M2
pen.text((0,pict_interval*3+unit_h*2.5),'M2',fill='black', font=method_font)
pen.text((method_interval+0.3*unit_w+unit_w*0,pict_interval*2+unit_h*2+20),'Strain map',fill='black', font=subtitle_font)
tmp = Image.open('M2_total.png')
canvas.paste(tmp,(method_interval+unit_w*0+pict_interval*0,unit_h*2+ pict_interval*3))

pen.text((method_interval+unit_w*(1+0.3)+pict_interval*1,pict_interval*2+unit_h*2+20),'M2 bond',fill='black', font=subtitle_font)
tmp = Image.open('M2_bond.png')
canvas.paste(tmp,(method_interval+unit_w*1+pict_interval*1,unit_h*2+ pict_interval*3))

pen.text((method_interval+unit_w*(2+0.3)+pict_interval*2,pict_interval*2+unit_h*2+20),'M2 angle',fill='black', font=subtitle_font)
tmp = Image.open('M2_angle.png')
canvas.paste(tmp,(method_interval+unit_w*2+pict_interval*2,unit_h*2+ pict_interval*3))
#Draw line
for n in range(1,3):
    pen.line((0,pict_interval*n+unit_h*n,unit_w*COL+ pict_interval*(COL-1)+method_interval,pict_interval*n+unit_h*n),fill='black',width=13)
'''

canvas.save(out_put_name,quality=save_quality)
display(Image.open(out_put_name))
```
