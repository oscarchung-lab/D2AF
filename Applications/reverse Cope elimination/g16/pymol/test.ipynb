```python
from PIL import Image, ImageDraw, ImageFont
import pandas as pd

# 打开图片
img1 = Image.open('M1_frag.png')
img2 = Image.open('M1_frag_show.png')

# 创建一个新的DataFrame
df = pd.DataFrame({'image1': [img1, img2]})
print(len(df))
print(len(df.columns))
# space for ltest
index_space = 150
col_space = 500

# 添加行头和列名
#indexs = ['M1', 'M2 total', 'M2 bond', 'M2 angle']
#cols = ['DFT', 'GFN2-xTB','ANI-2x','AIQM1']
indexs = ['M1', 'M2 total']
cols = ['DFT']
# 创建一个新的大图
new_img = Image.new('RGB', (img1.width + img2.width+index_space, img1.height*len(df)+col_space), (255, 255, 255))

# 将DataFrame中的图片粘贴到大图中
for i in range(len(df)):
    for j in range(len(df.columns)):
        new_img.paste(df.iloc[i, j], (col_space+j * df.iloc[i, j].width, index_space+i * df.iloc[i, j].height))

# 创建一个新的画布
draw = ImageDraw.Draw(new_img)

# 添加行头文字
font = ImageFont.truetype('arial.ttf', 99)
for i, index in enumerate(indexs):
    draw.text((10, (i+0.5) * img1.height ), str(index), font=font, fill=(0, 0, 0))

# 添加列名文字
for j, column in enumerate(cols):
    draw.text(((j+0.5) * img1.width, 10), str(column), font=font, fill=(0, 0, 0))
    
# 绘制表格
for i in range(len(df) + 1):
    draw.line((0, index_space + i * df.iloc[0, 0].height, img1.width + img2.width, index_space+i * df.iloc[0, 0].height), fill=(0, 0, 0), width=2)
for j in range(len(df.columns) + 1):
    draw.line((col_space + j * df.iloc[0, 0].width, 0, col_space + j * df.iloc[0, 0].width, max(img1.height, img2.height)), fill=(0, 0, 0), width=2)

# 显示大图
new_img.show()

# 将大图保存到本地
new_img.save('big_image.png')
```

```python
from PIL import Image, ImageDraw, ImageFont
import os
#method_dirs [g16 gfn2-xtb ani-2x aiqm1]
#methods [DFT GFN2-xTB ANI-2x AIQM1]
#fignames M1_M2 [M1_frag M2_total M2_bond M2_angle]
#fignames M3 [M3_total M3_bond M3_angle M3_frag]
method_space = 150
type_space = 500
img_width = 1200
img_height = 1200
font = ImageFont.truetype('timesbd.ttf', 99)

def Combine_Image_methods(fignames,figlabels,method_dirs,methodlabels):
    numdirs = len(method_dirs)
    new_img = Image.new('RGB', (img_width*numdirs + type_space, img_height*4 + method_space), (255, 255, 255))
    
    for i, dirtmp in enumerate(method_dirs):
        for j, figtmp in enumerate(fignames):
            img = Image.open(os.path.join(dirtmp,'pymol',figtmp))

            new_img.paste(img, (type_space+i * img_width, method_space+j * img_height))

    # 添加行头文字
    for i, figlabel in enumerate(figlabels):
        draw.text((10, (i+0.5) * img_height), figlabel, font=font, fill=(0, 0, 0))

    # 添加列名文字
    for j, methodlabel in enumerate(methodlabels):
        draw.text(((j+0.5) * img_width, 10), methodlabel, font=font, fill=(0, 0, 0))
    
    # 绘制表格
    for i in range(len(df) + 1):
        draw.line((0, method_space + i * img_height, type_space+img_width*numdirs, method_space+i * img_height), fill=(0, 0, 0), width=2)
    for j in range(len(df.columns) + 1):
        draw.line((type_space + j * img_width, 0, type_space + j * img_width, method_space+img_height*4), fill=(0, 0, 0), width=2)

    new_img.save('test.png')
    
    
  
figns = ['M1_frag', 'M2_total', 'M2_bond', 'M2_angle']
figls = ['M1 fragment', 'M2 total', 'M2 bond', 'M2 angle']
method_ds = ['g16','gfn2-xtb','ani-2x','aiqm1']
method_ls = ['DFT','GFN2-xTB','ANI-2x','AIQM1']

Combine_Image_methods(figns,figls,method_ds,method_ls)
    
```
